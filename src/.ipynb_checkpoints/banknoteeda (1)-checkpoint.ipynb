## 1. Imports

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats

# Consistent style
sns.set_theme(style='whitegrid', palette='muted')
plt.rcParams['figure.dpi'] = 120

print('Libraries loaded successfully.')
```

## 2. Load Dataset

```python
# Load dataset 
col_names = ['variance', 'skewness', 'curtosis', 'entropy', 'class']

df = pd.read_csv('data_banknote_authentication.txt', header=None, names=col_names)

print(f'Dataset shape: {df.shape}')
df.head()
```

## 3. Separate Features from Target

We separate `X` (features only) from `y` (ground truth class labels) **before any analysis**.  
This is critical for unsupervised learning — we **never feed the target into clustering**.

```python
# Drop target column
X = df.drop(columns=['class'])   # shape: (1372, 4)


y = df['class']     # labels(genuine vs forged)           

print(f'Feature matrix X: {X.shape}  →  columns: {list(X.columns)}')
print(f'Target vector  y: {y.shape}  →  classes: {y.unique()}')
print(f'Class distribution:\n{y.value_counts()}')
```

## 11. Skewness & Kurtosis Check

```python
skew_kurt = pd.DataFrame({
    'Skewness': X.skew(),
    'Kurtosis': X.kurtosis()
}).round(3)

print('Skewness & Kurtosis:')
print(skew_kurt)
```

## 12. Outlier Count (IQR Method)

```python
def count_outliers_iqr(df):
    Q1 = df.quantile(0.25)
    Q3 = df.quantile(0.75)
    IQR = Q3 - Q1
    outlier_mask = (df < (Q1 - 1.5 * IQR)) | (df > (Q3 + 1.5 * IQR))
    return outlier_mask.sum()

outlier_counts = count_outliers_iqr(X)
print('Outlier counts per feature (IQR method):')
print(outlier_counts)
print(f'\nTotal outliers: {outlier_counts.sum()} out of {len(X) * len(X.columns)} data points')
```

## 13. EDA Summary & Justification for Unsupervised Clustering

| Property | Value |
|---|---|
| Instances | 1,372 |
| Features | 4 (all continuous, numeric) |
| Missing values | 0 |
| Ground truth classes | 2 (genuine, forged) |
| Class balance | ~55% genuine, ~45% forged |

**Why this dataset suits unsupervised clustering:**
- All 4 features are continuous and numeric — no encoding needed.
- No missing values — no imputation required.
- The features show meaningful variance and some inter-feature correlation, making PCA useful for dimensionality reduction.
- Ground truth labels exist, enabling external metric evaluation (ARI, NMI) after clustering.
- The dataset is compact (1,372 rows) — ideal for comparing K-means, hierarchical, and DBSCAN without heavy compute.



```python
# ==========================================================
# CAPSTONE PREPROCESSING PIPELINE
# Using Existing Dataset Variable: df
# ==========================================================

# Import Libraries
import pandas as pd
import numpy as np

# Preprocessing
from sklearn.pipeline import Pipeline
from sklearn.compose import ColumnTransformer
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA

# ==========================================================
# STEP 1: COPY ORIGINAL DATA
# ==========================================================

# Preserve original dataset
data = df.copy()

```

```python
# ==========================================================
# STEP 2: REMOVE DUPLICATES
# ==========================================================

data = data.drop_duplicates()
```

```python

# ==========================================================
# STEP 3: REMOVE TARGET COLUMN
# ==========================================================
# Since clustering is unsupervised

X = data.drop('class', axis=1)
```

```python
# ==========================================================
# STEP 3.5: OUTLIER FLAGGING (NO REMOVAL)
# ==========================================================

# IQR method for outlier detection
Q1 = X.quantile(0.25)
Q3 = X.quantile(0.75)
IQR = Q3 - Q1

lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Flag rows with any outlier in any feature
X['outlier_flag'] = ((X < lower_bound) | (X > upper_bound)).any(axis=1)
```

```python
# ==========================================================
# STEP 4: IDENTIFY NUMERIC FEATURES
# ==========================================================

numeric_features = X.columns.tolist()
```

```python
# ==========================================================
# STEP 5: PREPROCESSING PIPELINE
# ==========================================================

numeric_pipeline = Pipeline(steps=[

    # Fill Missing Values
    ('imputer', SimpleImputer(strategy='median')),

    # Standardize Data
    ('scaler', StandardScaler())

])

# Column Transformer
preprocessor = ColumnTransformer(transformers=[

    ('num', numeric_pipeline, numeric_features)

])

```

```python
# ==========================================================
# STEP 6: CREATE NON-PCA DATASET
# ==========================================================

X_no_pca = preprocessor.fit_transform(X)

# Convert to DataFrame
df_no_pca = pd.DataFrame(
    X_no_pca,
    columns=numeric_features
)

```

```python
# ==========================================================
# STEP 7: CREATE PCA DATASET
# ==========================================================

pca = PCA(n_components=2)

X_with_pca = pca.fit_transform(X_no_pca)

# PCA DataFrame
df_with_pca = pd.DataFrame(
    X_with_pca,
    columns=['PC1', 'PC2']
)
```

```python
# ==========================================================
# STEP 8: STORE EXPLAINED VARIANCE
# ==========================================================

explained_variance = pca.explained_variance_ratio_

variance_summary = pd.DataFrame({
    'Principal_Component': ['PC1', 'PC2'],
    'Explained_Variance_Ratio': explained_variance
})
```

```python
# ==========================================================
# STEP 9: SAVE READY-TO-USE DATASETS
# ==========================================================

df_no_pca.to_csv('banknote_preprocessed_no_pca.csv', index=False)
df_with_pca.to_csv('banknote_preprocessed_with_pca.csv', index=False)

```

```python
# ==========================================================
# STEP 10: PRINT SUMMARY
# ==========================================================

print("="*50)
print("PREPROCESSING SUMMARY")
print("="*50)

print(f"\nOriginal Shape: {df.shape}")
print(f"After Cleaning: {data.shape}")

print("\nFeatures Used:")
print(numeric_features)

print("\nNo PCA Dataset Shape:")
print(df_no_pca.shape)

print("\nPCA Dataset Shape:")
print(df_with_pca.shape)

print("\nExplained Variance Ratio:")
print(variance_summary)

print("\nTotal Variance Retained:")
print(explained_variance.sum())

print("\nDatasets Saved Successfully.")
```
